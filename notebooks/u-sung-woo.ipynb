## 자료형 변경

```python
months = {
    "january":1,
    "february":2,
    "march":3,
    "april":4,
    "may":5,
    "june":6,
    "july":7,
    "august":8,
    "september":9,
    "october":10,
    "november":11,
    "december":12
}
```

```python
df_meteorshowers['bestmonth'] = df_meteorshowers['bestmonth'].map(months)
df_meteorshowers['startmonth'] = df_meteorshowers['startmonth'].map(months)
df_meteorshowers['endmonth'] = df_meteorshowers['endmonth'].map(months)
df_moonphases['month'] = df_moonphases['month'].map(months)
df_constellations['bestmonth'] = df_constellations['bestmonth'].map(months)

display(df_meteorshowers.head(2))
display(df_moonphases.head(2))
display(df_constellations.head(2))
```

## 날짜형 컬럼 만들기

```python
sdate = 2020*10000 + df_meteorshowers['startmonth']*100 + df_meteorshowers['startday']
sdate = pd.to_datetime(sdate, format='%Y%m%d')
df_meteorshowers['startdate'] = sdate
df_meteorshowers
```

```python
edate = 2020*10000 + df_meteorshowers['endmonth']*100 + df_meteorshowers['endday']
edate = pd.to_datetime(edate, format='%Y%m%d')
df_meteorshowers['enddate'] = edate
df_meteorshowers
```

```python
date = 2020*10000 + df_moonphases['month']*100 + df_moonphases['day']
date = pd.to_datetime(date, format='%Y%m%d')
df_moonphases['date'] = date
df_moonphases
```

```python
display(df_meteorshowers.head(2))
display(df_moonphases.head(2))
 
display(df_meteorshowers.info())
display(df_moonphases.info())
```

## 달의 위상을 숫자로 변경하여 파생컬럼 percentage 만들기

```python
#new moon= 0, first quarter= 0.5, full moon= 1, third quarter= 0.5

moonphase_to_percentage = {
    'new moon': 0,
    'first quarter': 0.5,
    'full moon': 1.0,
    'third quarter': 0.5,
    pd.NA: pd.NA}
df_moonphases['percentage'] = df_moonphases['moonphase'].map(moonphase_to_percentage)
df_moonphases
```

```python
df_meteorshowers.drop(['startmonth', 'startday', 'endmonth', 'endday', 'hemisphere'], axis=1)
```

```python
df_constellations.head()
```

```python
del_colums = 'besttime'
df_constellations = df_constellations.drop(del_colums, axis=1)
df_constellations
```

```python
df_moonphases
```

```python
del_colums = ['month', 'day', 'moonphase', 'specialevent']
df_moonphases = df_moonphases.drop(del_colums, axis=1)
df_moonphases
```

```python
lastphase = 0
for idx, data in df_moonphases.iterrows():
        print(idx)
        print(data['percentage'])
        print(pd.isnull(data['percentage']))
        if pd.isnull(data['percentage']):
                df_moonphases.at[idx,'percentage'] = lastphase
        else:
                lastphase = data['percentage']
df_moonphases 
```
